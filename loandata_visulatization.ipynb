#### Imported python packages

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Loaded data using pandas

```python
data_frame = pd.read_csv("/content/loan_data_set.csv")
```

#### Get head data

```python
data_frame.head()
```

#### Get shape

```python
data_frame.shape
```

#### Visualization Married data

```python
sns.countplot(x="Married", data=data_frame)
data_frame["Married"].value_counts()
```

#### Visualization Self Employeed Data

```python
sns.countplot(x="Self_Employed", data=data_frame)
data_frame["Self_Employed"].value_counts()
```

#### Visualization Education data

```python
sns.countplot(x="Education", data=data_frame)
data_frame["Education"].value_counts()
```

#### Visualization of Property Area using pie plot

```python
data_frame["Property_Area"].value_counts().plot(kind="pie" ,autopct="%.2f")
```

#### Visualization of Applicant Income

```python
plt.hist(data_frame["ApplicantIncome"], bins=10, rwidth=0.95)
```

#### Visualization of Load Amount

```python
plt.hist(data_frame["LoanAmount"], bins=10, rwidth=0.95)
```

#### Visualization of Applicant Income using displot

```python
sns.distplot(data_frame["ApplicantIncome"], bins=10, kde=True)
```

#### Visualization of Loan Amount using boxplot

```python
sns.boxplot(data_frame['LoanAmount'])
```

#### Visualization of Amount using Violinplot

```python
sns.violinplot(x="LoanAmount", data=data_frame)
```

#### Visualization of Applicant Income using Scatterplot

```python
sns.scatterplot(data=data_frame, x="ApplicantIncome", y="LoanAmount")
```

#### Visualization using scatterplot of multiple of columns

```python
sns.scatterplot(data=data_frame, x="ApplicantIncome", y="LoanAmount", hue="Gender", style="Self_Employed", size="Property_Area")
```

#### Visualization of Property Area and Applicant Income using barplot

```python
sns.barplot(data=data_frame, x="Property_Area", y="ApplicantIncome")
```

#### Visualization of Property area and applicant income with gender

```python
sns.barplot(data=data_frame, x="Property_Area", y="ApplicantIncome", hue="Gender")
```

```python
#### Visualization of identifying Gender depening on Loan Amount
```

```python
sns.boxplot(data=data_frame, x="Gender", y="LoanAmount")
```

#### Visualization of Gender and Loan Amount according to how many of them are self employed

```python
sns.boxplot(data=data_frame, x="Gender", y="LoanAmount", hue="Self_Employed")
```

```python
sns.distplot(data_frame[data_frame["Property_Area"]=="Urban"]["LoanAmount"], hist=False,color="red")
sns.distplot(data_frame[data_frame["Property_Area"]=="Rural"]["LoanAmount"], hist=False,color="black")
sns.distplot(data_frame[data_frame["Property_Area"]=="Semiurban"]["LoanAmount"], hist=False)
```

#### Visualization using heatmap of Property area and self employed data

```python
sns.heatmap(pd.crosstab(data_frame["Property_Area"], data_frame["Self_Employed"]))
```

#### Visualization of whole dataframe using pairplot

```python
sns.pairplot(data_frame)
```
